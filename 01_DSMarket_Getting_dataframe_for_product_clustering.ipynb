A continuación voy a quedarme solo con la información necesaria para el clustering de productos. En este sentido no quiero informaciones cerca de las tiendas o de la region, en cuanto quiero clasificar los productos solamente en base a sus proprias caracteristicas.

```python
df.drop([
  'id',
  'store',
  'store_code',
  'region',
  'revenue'],
        axis=1,inplace=True)
```

Me quedo solo con los años completos: 2012-2015

```python
df = df[(df['date']>='2011-01-29')&(df['date']<='2016-01-28')]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
# agrupo las ventas de cada item
df['units_sold'] = df.groupby(['item','date'])['units_sold'].transform(np.sum)
# calculo el precio de cada item como el promedio entre los precios de cada tienda
df['price_per_unit'] = df.groupby(['item','date'])['price_per_unit'].transform(np.mean)
```

```python
df = df.sort_values(['item','date'])
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df = df.drop_duplicates()
```
